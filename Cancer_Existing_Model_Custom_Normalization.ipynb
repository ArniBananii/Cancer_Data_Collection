```python
"""
Random Seet setzen, damit Ergebnisse reproduzierbar sind
"""

import random
def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

def seed_worker(worker_id):
    worker_seed = torch.initial_seed() % 2**32
    np.random.seed(worker_seed)
    random.seed(worker_seed)

set_seed(42)

g = torch.Generator()
g.manual_seed(42)
```

```python
"""
Creates new images in the Directory "AugmentedData/Train - these new images are stain normalized
"""

import torch
from torchvision.io import read_image, write_jpeg  # or write_png
import os
from torch_staintools import normalizer as torch_staintools

# Set up directories
base_input_dir = "./Data/Train"
normalized_dir = "./AugmentedData/Train/"
augmentation_images_dir = "./stain_bilder"
os.makedirs(normalized_dir, exist_ok=True)


# Normalize and save each image
for root, dirs, files in os.walk(base_input_dir):
    for dir in dirs:

        # Load target image for normalization
        dir_files = os.listdir(os.path.join(base_input_dir, dir))
        normalization_img_file_name = dir + '.jpg'
        target_img = read_image(os.path.join(augmentation_images_dir, normalization_img_file_name)).float() / 255.0
        target_img = target_img.unsqueeze(0)
        normalizer = torch_staintools.NormalizerBuilder.build('macenko')
        normalizer.fit(target_img)
        normalizer.fit(target_img)
        for fname in dir_files:
            if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                img = read_image(os.path.join("".join([base_input_dir, '/', dir]), fname)).float() / 255.0
                img = img.unsqueeze(0)
                norm_img = normalizer.transform(img)
                # Convert back to uint8 for saving
                norm_img_uint8 = (norm_img * 255).clamp(0, 255).byte()

                if len(norm_img_uint8.shape) == 2:
                    norm_img_uint8 = norm_img_uint8.unsqueeze(-1).repeat(1, 1, 3)
                elif len(norm_img_uint8.shape) == 4:
                    norm_img_uint8 = norm_img_uint8.squeeze(0)
                elif norm_img_uint8.shape[-1] != 3:
                    norm_img_uint8 = norm_img_uint8.permute(1, 2, 0)

                os.makedirs(os.path.join(normalized_dir, dir), exist_ok=True)
                write_jpeg(norm_img_uint8, os.path.join(normalized_dir, dir, fname))
```

```python
import Augmentor
import random
from torchvision.io import read_image, write_jpeg  # or write_png
from torch_staintools import normalizer as torch_staintools


class CustomDataset(Dataset):
    def __init__(self, data_dir, transform=None, augment_times=1):
        self.data_dir = data_dir
        self.transform = transform
        self.augment_times = augment_times

        self.samples = []  # list of (image_path, class_name)
        self.classes = sorted(os.listdir(data_dir))
        self.class_to_idx = {cls_name: idx for idx, cls_name in enumerate(self.classes)}

        # scan all image paths
        for cls in self.classes:
            cls_path = os.path.join(data_dir, cls)
            for fname in os.listdir(cls_path):
                if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                    self.samples.append((os.path.join(cls_path, fname), cls))

        target_img = read_image("stain_normalized_image_1.jpg")
        target_img = target_img.unsqueeze(0)
        normalizer = torch_staintools.NormalizerBuilder.build('macenko')
        #normalizer.fit(target_img)
        #self.normalizer = normalizer

        self.augmentors = {}
        for cls in self.classes:
            aug = Augmentor.Pipeline()
            aug.rotate(probability=0.7, max_left_rotation=10, max_right_rotation=10)
            aug.flip_left_right(probability=0.5)
            aug.zoom(probability=0.5, min_factor=1.1, max_factor=1.3)
            aug.random_contrast(probability=0.3, min_factor=0.7, max_factor=1.3)
            aug.random_brightness(probability=0.3, min_factor=0.7, max_factor=1.3)
            self.augmentors[cls] = aug


    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, cls = self.samples[idx]
        label = self.class_to_idx[cls]

        img = read_image(img_path).float() / 255.0
        #img = img.unsqueeze(0)

        # stain normalisieren
        #norm_img = self.normalizer.transform(img)
        #norm_img = norm_img.squeeze(0)

        # PIL Image f체r Augmentor
        img_pil = transforms.ToPILImage()(img)

        augmented_imgs = [img_pil]
        if self.augment_times > 1:
            p = self.augmentors[cls]
            for _ in range(self.augment_times - 1):
                aug_img = p.sample(1)[0]
                augmented_imgs.append(aug_img)

        if self.transform:
            augmented_imgs = [self.transform(im) for im in augmented_imgs]
        else:
            augmented_imgs = [transforms.ToTensor()(im) for im in augmented_imgs]

        return augmented_imgs[0], label
```

```python
"""
Nun definieren wir Transformatoren f체r die Daten, laden das Dataset und teilen diese in Trainings- und Validierungs채tze auf:
"""
# Definieren Sie Transformationen f체r das Training und die Validierung
transform = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),

        #transforms.AutoAugment(AutoAugmentPolicy.CIFAR10),
        transforms.ToTensor(),
        #transforms.Normalize([0.485, 0.456, 0.406],
                             #[0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        #transforms.Normalize([0.485, 0.456, 0.406],
                            #[0.229, 0.224, 0.225])
    ])
}

train_dataset = torchvision.datasets.ImageFolder(root='AugmentedData/Train', transform=transform["train"])
test_dataset = torchvision.datasets.ImageFolder(root='AugmentedData/Test', transform=transform["test"])

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=64, shuffle=True, num_workers=0, pin_memory=True, generator=g, worker_init_fn=seed_worker)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=64, shuffle=False, num_workers=0, pin_memory=True, generator=g, worker_init_fn=seed_worker)
```

```python
# Device configuration, Pr체fe ob GPU verf체gbar ist, falls ja, lasse das Modell auf eie GPU laufen, ansonsten CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# W채hle ein Modell aus unter: https://pytorch.org/vision/main/models.html
model = models.mobilenet_v2(weights=models.MobileNet_V2_Weights.DEFAULT)


# mit diesem Code snippet k철nnen wir die Parmeter freeze, sodass diese nicht gelernt werden. Ziel Wir trainieren nur den letzten layer, den wir am Ende hinzuf체gen.
"""
# freeze all model parameters
for param in model.parameters():
    param.requires_grad = False
"""
#num_ftrs = model.heads[0].in_features
#model.heads[0] = torch.nn.Linear(num_ftrs, len(train_dataset.classes))



model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
"""
Trainieren des Modells
"""
# Trainingsschleife
num_epochs = 3
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    pbar = tqdm(train_loader)
    nr_samples = 0
    labels_global = []
    preds_global = []
    for idx, (images, labels) in enumerate(pbar, start=1):
        # Vorhersage und Verlust berechnen
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        loss = criterion(outputs, labels)

        # Gradienten zur체cksetzen, Backwardpropagation und Optimierung
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        nr_samples += images.shape[0]
        running_loss += loss.item() * images.size(0)
        running_corrects += torch.sum(preds == labels)
        labels_global.extend(labels.tolist())
        preds_global.extend(preds.tolist())
        cnt_acc = running_corrects/nr_samples
        cnt_loss = running_loss/nr_samples
        f1 = f1_score(labels_global, preds_global, average='micro')
        pbar.set_description(f'Epoch [{epoch+1}/{num_epochs}], Loss: {cnt_loss:.3f}, Acc: {cnt_acc:.3f}, F1-Score: {f1:.3f}')
```

```python
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=1, shuffle=False)
errors = []
model.eval()
running_corrects = 0
nr_samples = 0
for images, labels in tqdm(test_loader):
    images, labels = images.to(device), labels.to(device)
    outputs = model(images)
    _, preds = torch.max(outputs.data, 1)
    running_corrects += torch.sum(preds == labels)
    nr_samples += images.shape[0]
    if preds != labels:
      errors.append({"image":images.to("cpu")[0], "y_hat":preds.to("cpu")[0].item(), "label":labels.to("cpu")[0].item()})
print()
print(f"Accuracy on Test dataset: {(running_corrects/nr_samples)*100:.3f} %")
```
